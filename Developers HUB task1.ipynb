**Problem Statement:**

Understanding a dataset's structure, distribution, and the relationships between its variables is a fundamental prerequisite for any machine learning project. Without proper exploration, it is difficult to identify trends, potential outliers, or feature dependencies that impact model performance.

**Goal:**

The objective of this task is to perform an Exploratory Data Analysis (EDA) on the Iris dataset. This includes loading the data, inspecting its features, generating summary statistics, and creating visualizations to understand the distributions and relationships between the variables

```python
# Import necessary libraries for data handling and visualization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset directly from seaborn's built-in repository
df = sns.load_dataset('iris')

# Display the first 10 rows
print("First 10 rows of the dataset:")
print(df.head(10))
```

```python
# Display the dimensions of the dataframe and the list of column headers
print(f"Shape of the dataset: {df.shape}")
print(f"Columns: {df.columns.tolist()}")

# Provide detailed technical information about the dataset (data types, memory, non-null counts)
print("\nDataset Info:")
df.info()
# Generate descriptive statistics (mean, std, min, max, quartiles) for all numerical columns
print("\nSummary Statistics:")
print(df.describe())
```

```python
# 1. Scatter Plot (Relationship between Sepal Length and Sepal Width)
# Creating a scatter plot to analyze the correlation between sepal length and width,
# colored by the species to see if they form distinct clusters.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()

# 2. Histogram (Distribution of Petal Length)
# Use a histogram with a Kernel Density Estimate (KDE) curve to view the
# distribution patterns of petal length.
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='petal_length', kde=True, color='purple')
plt.title('Distribution of Petal Length')
plt.show()

# 3. Box Plot (Identify outliers for each feature by species)
# Generate box plots to visualize the range of sepal length per species
# and identify potential outliers (indicated by points outside the whiskers).
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Sepal Length by Species')
plt.show()
```
